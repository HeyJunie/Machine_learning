```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split

fetures = iris.data[:, 2:]
labels = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(fetures, labels,
                                                    test_size=0.2, 
                                                    stratify=labels,
                                                    random_state=13)
```

```python
X_train.shape, X_test.shape
```

```python
import numpy as np

np.unique(y_test, return_counts=True)
```

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
plot_tree(iris_tree);
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(iris.data[:, 2:])
accuracy_score(iris.target, y_pred_tr)
```

```python
from mlxtend.plotting import plot_decision_regions
plt.figure(figsize=(14, 8))
plot_decision_regions(X=X_train, y=y_train, clf=iris_tree, legend=2)
plt.show()
```

```python
y_pred_test = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred_test)
```

```python
scatter_highlight_kwargs = {'s':150, 'label':'Test_data', 'alpha':0.9}
scatter_kwargs = {'s':120, 'edgecolor':None, 'alpha':0.9}
plt.figure(figsize=(12, 8))
plot_decision_regions(X=fetures, y=labels, clf=iris_tree, legend=2,
                     X_highlight=X_test, scatter_highlight_kwargs=scatter_highlight_kwargs,
                     scatter_kwargs=scatter_kwargs,
                     contourf_kwargs={'alpha':0.2})
```
